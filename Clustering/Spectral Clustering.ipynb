```python
import pandas as pd
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
```

```python
dataset.head(21)
```

```python

X= dataset.iloc[:,[3,4]].values
X
```

```python
from sklearn.cluster import SpectralClustering
sc =SpectralClustering(n_clusters=5, eigen_solver=None, n_components=None, random_state=None, n_init=10)
sc.fit_predict(X)  
spectral_cluster =sc.fit_predict(X)
```

```python
spectral_cluster
```

```python
sc_cluster = pd.DataFrame(dataset)
sc_cluster.head()
```

```python
sc_cluster['sc_Cluster_Group'] = spectral_cluster
```

```python
sc_cluster
```

```python
sc_cluster.to_csv("sc_cluster.csv", index=False)
```

```python
import seaborn as sns

facet =sns.lmplot(data=sc_cluster, x=sc_cluster.columns[3], y=sc_cluster.columns[4], hue = sc_cluster.columns[5], fit_reg=False, legend=True, legend_out=True)

```
